```python
countries = CAT.columns[1:]
CAT[countries] = CAT[countries].apply(lambda x: x* 1000)
CAT.head()
```

```python
CAT["Month/year_error"] = CAT["Month/year"].apply(lambda x: "17" in x)
```

```python
CAT.head()
```

```python
CAT_2017 = CAT[CAT["Month/year_error"]==True]    
CAT_2017 = CAT_2017.drop(["Month/year_error"], axis=1)

CAT_2017.to_csv(".\clean_datasets\CAT_2017.csv") 
```

```python
totals =[]
for i in CAT_2017:
    totals.append(CAT_2017[i].sum())


total_values = totals[2:]

CAT_2017.columns

col = ["Alemanya", 'Bèlgica', 'França', 'Irlanda',
       'Itàlia', 'Països Baixos', 'Nordic countries', 'Portugal', 'Regne Unit',
       'Rússia', 'Suïssa', "Resta d'Europa", 'Estats Units', "Rest d'America",
       'Resta del món']
  

total_2017 = pd.DataFrame({"total_values":total_values, "Country":col})
```
